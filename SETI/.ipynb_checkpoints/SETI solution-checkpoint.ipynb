```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
```

```python
fir_taps = 1445
x = np.fromfile('polarization.f32', 'float32')
```

```python
corr = scipy.signal.correlate(x, x, mode='full')
```

```python
plt.plot(corr)
```

```python
plt.plot(corr[corr.size//2:])
```

```python
plt.plot(corr[corr.size//2+20:])
```

```python
plt.plot(corr[corr.size//2+20:][:100000])
```

```python
plt.plot(corr[corr.size//2+30000:][:10000])
a = np.argmax(corr[corr.size//2+30000:][:10000])
plt.plot(a, corr[corr.size//2+30000:][a], 'o')
a = a + 30000
```

```python
z = x[:x.size//a*a].reshape(-1, a)
plt.figure(figsize=(10, 10))
plt.imshow(z, aspect='auto')
```

```python
b = a #- 4
z = x[:x.size//b*b].reshape(-1, b)
plt.figure(figsize=(10, 10))
plt.imshow(z, aspect='auto')
```

```python
z = x[:x.size//b*b].reshape(-1, b)
plt.figure(figsize=(10, 10))
plt.imshow(z.T, aspect='auto', cmap='gray')
```

```python
z = x[:x.size//b*b].reshape(-1, b)
plt.figure(figsize=(10, 3))
plt.imshow(np.roll(z.T, -12000, axis=0),
           aspect='auto', cmap='gray',
           vmin=-0.05, vmax=0.05)
```

```python
fir_taps_pwr = 579
fir_taps_pwr2 = 14455
delay_difference = (fir_taps_pwr-1) // 2 + (fir_taps_pwr2-1) // 2 - (fir_taps-1) // 2
pwr = np.fromfile('power.f32', 'float32')[delay_difference:]
pwr /= np.average(pwr)
plt.plot(pwr[:100000])
```

```python
w = x[:pwr.size]/pwr
w = w[:w.size//b*b].reshape(-1, b)
plt.figure(figsize=(10, 3))
plt.imshow(np.roll(w.T, -12000, axis=0),
           aspect='auto', cmap='gray',
           vmin=-0.05, vmax=0.05)
```
