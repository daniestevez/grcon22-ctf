```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('hcro2.png')[..., 0]
plt.imshow(img, cmap='gray')
```

```python
spectrum = np.zeros((2048, 2048), 'complex')
spectrum[512:-512, 500:1532] = img
spectrum *= np.exp(1j * 2 * np.pi * np.random.uniform(size=spectrum.shape))
```

```python
x = np.fft.ifft(np.fft.fftshift(spectrum, axes=1))
y = np.tile(x, (1, 16))
y += 0.005 * (np.random.randn(*y.shape) + 1j * np.random.randn(*y.shape))
```

```python
y.ravel().astype('complex64').tofile('painting.c64')
```
